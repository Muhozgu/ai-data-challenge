```
"""
EY 2026 AI & Data Challenge — Water Quality Prediction Model v6
Author: Guy-Brillant Muhoza
Score: 0.274 (best submission so far)

Model: GradientBoostingRegressor with spatial holdout validation
Key features: spectral indices, cyclical time encoding, PET interactions,
              SA city distances, log-transform for DRP target
"""

import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')

# ── 1. Load Data ───────────────────────────────────────────────────────────
wq     = pd.read_csv('water_quality_training_dataset.csv')
tc_tr  = pd.read_csv('terraclimate_features_training.csv')
tc_val = pd.read_csv('terraclimate_features_validation.csv')
ls_tr  = pd.read_csv('landsat_features_training.csv')
ls_val = pd.read_csv('landsat_features_validation.csv')

keys  = ['Latitude', 'Longitude', 'Sample Date']
train = wq.merge(tc_tr, on=keys, how='left').merge(ls_tr, on=keys, how='left')
val   = tc_val.merge(ls_val, on=keys, how='left')

# ── 2. Feature Engineering ─────────────────────────────────────────────────
def add_features(df):
    df = df.copy()
    df['_date']     = pd.to_datetime(df['Sample Date'], dayfirst=True)
    df['month']     = df['_date'].dt.month
    df['year']      = df['_date'].dt.year
    df['doy']       = df['_date'].dt.dayofyear

    # Cyclical time encoding
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
    df['doy_sin']   = np.sin(2 * np.pi * df['doy'] / 365)
    df['doy_cos']   = np.cos(2 * np.pi * df['doy'] / 365)
    df['season']    = df['month'] % 12 // 3

    # SA wet season flag (Oct–Mar)
    df['wet_season'] = df['month'].isin([10, 11, 12, 1, 2, 3]).astype(int)

    # Landsat spectral indices
    nir    = df.get('nir',    pd.Series(np.nan, index=df.index))
    green  = df.get('green',  pd.Series(np.nan, index=df.index))
    swir16 = df.get('swir16', pd.Series(np.nan, index=df.index))
    swir22 = df.get('swir22', pd.Series(np.nan, index=df.index))

    df['NDVI']       = (nir - swir16)  / (nir + swir16  + 1e-9)
    df['NIR_GREEN']  = nir / (green   + 1e-9)
    df['SWIR_ratio'] = swir16 / (swir22 + 1e-9)

    # TerraClimate PET interactions
    if 'pet' in df.columns:
        df['pet_sin'] = df['pet'] * df['month_sin']
        df['pet_cos'] = df['pet'] * df['month_cos']

    return df.drop(columns=['_date', 'month', 'doy'])

train = add_features(train)
val   = add_features(val)

# ── 3. Feature Columns ─────────────────────────────────────────────────────
FEAT_COLS = [c for c in [
    'Latitude', 'Longitude',
    'month_sin', 'month_cos', 'doy_sin', 'doy_cos', 'year', 'season', 'wet_season',
    'pet', 'pet_sin', 'pet_cos',
    'nir', 'green', 'swir16', 'swir22', 'NDMI', 'MNDWI',
    'NDVI', 'NIR_GREEN', 'SWIR_ratio'
] if c in train.columns]

print(f"Using {len(FEAT_COLS)} features: {FEAT_COLS}")

TARGETS = ['Total Alkalinity', 'Electrical Conductance', 'Dissolved Reactive Phosphorus']

X_train = train[FEAT_COLS]
X_val   = val[FEAT_COLS]

# ── 4. Spatial Holdout Validation ──────────────────────────────────────────
# Hold out 20% of unique locations to simulate generalization to new regions
unique_locs = train[['Latitude', 'Longitude']].drop_duplicates().sample(
    frac=1, random_state=99).reset_index(drop=True)
n = len(unique_locs)
holdout_set = set(zip(
    unique_locs.iloc[:int(n * 0.2)]['Latitude'],
    unique_locs.iloc[:int(n * 0.2)]['Longitude']
))
sp_val_mask = train.apply(
    lambda r: (r['Latitude'], r['Longitude']) in holdout_set, axis=1)
sp_tr_mask  = ~sp_val_mask

print(f"\nSpatial holdout — train: {sp_tr_mask.sum()}, holdout: {sp_val_mask.sum()} rows")

# ── 5. Train & Predict ─────────────────────────────────────────────────────
print("\n--- Spatial holdout R² ---")
preds_final = {}

for target in TARGETS:
    y       = train[target]
    use_log = (target == 'Dissolved Reactive Phosphorus')

    pipe = Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler',  RobustScaler()),
        ('model',   GradientBoostingRegressor(
            n_estimators=200,
            learning_rate=0.05,
            max_depth=4,
            subsample=0.8,
            min_samples_leaf=8,
            random_state=42
        ))
    ])

    # Spatial holdout evaluation
    y_tr  = train.loc[sp_tr_mask,  target]
    y_val = train.loc[sp_val_mask, target]

    if use_log:
        pipe.fit(X_train[sp_tr_mask], np.log1p(y_tr))
        sp_preds = np.expm1(pipe.predict(X_train[sp_val_mask]))
    else:
        pipe.fit(X_train[sp_tr_mask], y_tr)
        sp_preds = pipe.predict(X_train[sp_val_mask])

    sp_r2 = r2_score(y_val, sp_preds)
    print(f"  {target}: Spatial holdout R²={sp_r2:.4f}")

    # Final model trained on ALL training data
    if use_log:
        pipe.fit(X_train, np.log1p(y.clip(lower=0)))
        preds_final[target] = np.expm1(pipe.predict(X_val))
    else:
        pipe.fit(X_train, y)
        preds_final[target] = pipe.predict(X_val)

    train_r2 = r2_score(y, np.expm1(pipe.predict(X_train)) if use_log else pipe.predict(X_train))
    print(f"  {target}: Full train R²={train_r2:.4f}")

# ── 6. Build Submission ────────────────────────────────────────────────────
submission = val[keys].copy()
submission['Total Alkalinity']              = np.clip(preds_final['Total Alkalinity'], 0, 600)
submission['Electrical Conductance']        = np.clip(preds_final['Electrical Conductance'], 0, 5000)
submission['Dissolved Reactive Phosphorus'] = np.clip(preds_final['Dissolved Reactive Phosphorus'], 0, 2000)

submission.to_csv('submission_EY2026.csv', index=False)
print(f"\nSubmission saved! Shape: {submission.shape}")
print(submission.head())
```
